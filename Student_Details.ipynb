```python
# Student result project
# Uses of this project
# A Teacher can Input:
# 1. Add students
# 2. View all students
# 3. Check Results
# 4. Exit

# Dictionary-To store data
# Loops-Menu system
# conditional statement-logic


```

```python
# @title
student = {}

while True:
  print("\n-----STUDENT MANAGER APP-----")
  print("1. Add Student")
  print("2. View Student")
  print("3. Result Student")
  print("4. Exit Student")

  choice = input("Enter your choice: ")
  if choice=='1':
    name=input("Enter student name: ")
    mark=int(input("Enter marks: "))
    student[name]=mark
    print(f"{name} Student added successfully!")
  elif choice=="2":
    if not student:
      print("No student found!")
    else:
      for name, marks in student.items():
        print(name, ":", marks)
  elif choice=="3":
    name=input("Enter student name: ")
    if name in student:
      marks=student[name]

      if marks>=40:
        print("Pass")
      else:
        print("Fail")
    else:
      print("Student name not found")
  elif choice == "4":
    print("Exiting...")
    break
  else:
    print("Invalid Input")
```
